```
# AND gate using nn 
import numpy as np 
```

```
#Input datasets
inputs = np.array([[0,0],[0,1],[1,0],[1,1]])
expected_output = np.array([[0],[0],[0],[1]])
```

```
inputs.shape
expected_output.shape
```

```
epochs = 5000
lr = 0.1
inputLayerNeurons, outputLayerNeurons = 2,1
# For AND gate we don't need hidden layer
# we can use hidden layer but it will be useless
```

```
#Random weights and bias initialization
seed=10
output_weights = np.random.uniform(size=(inputLayerNeurons,outputLayerNeurons))
output_bias = np.random.uniform(size=(1,outputLayerNeurons))

print("Initial output weights: ",end='')
print(*output_weights)
print("Initial output biases: ",end='')
print(*output_bias)


```

```
output_weights.shape
```

```
def sigmoid (x):
    return 1/(1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)
```

```
#Training algorithm
for _ in range(epochs):

	output_layer_activation = np.dot(inputs,output_weights)
	output_layer_activation += output_bias
	predicted_output = sigmoid(output_layer_activation)
 
  #Backpropagation
	error = expected_output - predicted_output
	d_predicted_output = error * sigmoid_derivative(predicted_output)

	#Updating Weights and Biases
	output_weights += inputs.T.dot(d_predicted_output) * lr
	output_bias += np.sum(d_predicted_output,axis=0,keepdims=True) * lr
	print("output weights: ",end='')
	print(output_weights)


	
```

```
print("Final output weights: ", *output_weights)
print("Final output bias: ", *output_bias)

print("\nOutput from neural network after 5,000 epochs: ",end='')
print(*predicted_output)
```
